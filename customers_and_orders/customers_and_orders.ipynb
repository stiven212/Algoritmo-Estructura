**Question 5:** What is the most common last name?

```python
from os.path import exists
import csv

# check if file can be found and open the file
datafile = "./sample_data/customers.csv"
if not exists(datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")

# open file and init vars
# evaluate the contents of the file
surnames = []

with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')
  for state in csvreader:
    #print(state[4])
    surnames.append(state[2].strip().upper())
  #print(surnames)

uniqueSurnames = set(surnames)
#print(uniqueSurnames)

surnamesNumber = {}
for us in uniqueSurnames:
  surnamesRepeated = 0
  for sur in surnames:
    if us == sur: surnamesRepeated += 1
  surnamesNumber[us]=surnamesRepeated

#print(surnamesNumber)

surnamesNumberSorted = dict(sorted(surnamesNumber.items(), key=lambda item: item[1], reverse=True))
print("The most common last name are",surnamesNumberSorted)
```

```python
# open and evaluate orders.csv

from os.path import exists
import csv
import datetime
# check if file can be found and open the file
datafile = "./sample_data/orders.csv"
if not exists(datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")

# open file and init vars
# evaluate the contents of the file
ordersCode = []
with open(datafile, 'r') as fl:
  orders = csv.reader(fl, delimiter=',')
  for order in orders:
    #print(order)
    ordersCode.append(order[1])
  #print(ordersCode)

uniqueOrders = set(ordersCode)
#print(uniqueOrders)
print("Number of unique orders",len(uniqueOrders)-1)

#Average items per order
itemsPerOrder = {}
for uo in uniqueOrders:
  item = 0
  for order in ordersCode:
    if uo == order:
      item += 1
  itemsPerOrder[uo]=item

print(itemsPerOrder)

itemTotal = 0
for key, value in itemsPerOrder.items():
  itemTotal += value

print("Average items per order",round(itemTotal/len(uniqueOrders),2))

#Highest number of items per order
itemsPerOrderSorted = dict(sorted(itemsPerOrder.items(), key=lambda item: item[1], reverse=True))
print("Highest number of items per order",itemsPerOrderSorted)

#Number of orders in October 2021
octoberOrders = []

with open(datafile, 'r') as fl:
  orderList = csv.reader(fl, delimiter=',')
  for ol in orderList:
    if len(ol[2]) != 0:
      if ol[2] != 'Date':
        date = datetime.datetime.strptime(ol[2], '%Y-%m-%d %H:%M:%S.%f')
        if date.month == 10:
          if date.year == 2021:
            octoberOrders.append(ol[1])

octoberOrders = set(octoberOrders)
#print(octoberOrders)

print("Number of orders in October 2021",len(octoberOrders))


#Customers who spent more in 2021
customerCode = {}
with open(datafile, 'r') as fl:
  customers = csv.reader(fl, delimiter=',')
  for custom in customers:
    if len(custom[2]) != 0:
      if custom[2] != 'Date':
        date = datetime.datetime.strptime(custom[2], '%Y-%m-%d %H:%M:%S.%f')
        if date.year == 2021:
          if custom[0] not in customerCode:
            customerCode[custom[0]] = float(custom[5])
          else:
            customerCode[custom[0]] += float(custom[5])

customerCodeSorted =  dict(sorted(customerCode.items(), key=lambda item: item[1], reverse=True))
print("Customers who spents more in 2021",customerCodeSorted)

  #print(ordersCode)

```

```python
#Best month for sales
months = {}
with open(datafile, 'r') as fl:
  sells = csv.reader(fl, delimiter=',')
  for sell in sells:
    if len(sell[2]) != 0:
      if sell[2] != 'Date':
        date = datetime.datetime.strptime(sell[2], '%Y-%m-%d %H:%M:%S.%f')
        month = date.month
        #print(month)
        if month not in months:
          months[month] = float(sell[5])
        else:
          months[month] += float(sell[5])


monthsSorted =  dict(sorted(months.items(), key=lambda item: item[1], reverse=True))
print("Best month for sales",monthsSorted)
```
